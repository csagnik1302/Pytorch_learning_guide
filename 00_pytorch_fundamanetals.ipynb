## Introduction to Tensors

### Tensors are a way to represent any form of structured or unstructured data in a multidimensional numeric form



### Creating Tensors - Pytorch Tensors are created using ```torch.tensor()``` command. The argument that it takes can be of upto any dimension

````python
scalar=torch.tensor(7)  # Creates a Scalar (A Rank 0 Scalar) with stored numeric value 7
scalar
````

### Tensor Attributes

````python
scalar.ndim     # Shows the Dimension of the Tensor
````
